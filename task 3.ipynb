```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

np.random.seed(10)
temps = np.random.randint(25, 100, 8)  
temps.sort() 
sales = [30, 40, 50, 60, 70, 80, 90, 100]  

data = pd.DataFrame({
    "Temperature": temps,
    "Sales": sales
})


X = data[["Temperature"]] 
y = data["Sales"]          
model = LinearRegression()
model.fit(X, y)

print("Predicted Sales for 85 degrees:", model.predict([[85]]))
print("Predicted Sales for 40 degrees:", model.predict([[40]]))
```
