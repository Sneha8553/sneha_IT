```python

import pandas as pd
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='study_hours_per_day', y='exam_score', hue='gender', palette='coolwarm')
plt.title('Study Hours vs Exam Score (Colored by Gender)', fontsize=14)
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.legend(title='Gender')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='gender', y='exam_score', palette='pastel')
plt.title('Average Exam Score by Gender')
plt.ylabel('Average Exam Score')
plt.xlabel('Gender')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
bins = [0, 1, 3, 5, 10]
labels = ['0-1 hrs', '1-3 hrs', '3-5 hrs', '5+ hrs']
df['sm_usage_group'] = pd.cut(df['social_media_hours'], bins=bins, labels=labels)

sm_counts = df['sm_usage_group'].value_counts().sort_index()

plt.figure(figsize=(7,7))
plt.pie(sm_counts, labels=sm_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Social Media Usage Distribution')
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['sleep_hours'], bins=8, kde=True, color='skyblue')
plt.title('Distribution of Sleep Hours')
plt.xlabel('Sleep Hours per Day')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```
